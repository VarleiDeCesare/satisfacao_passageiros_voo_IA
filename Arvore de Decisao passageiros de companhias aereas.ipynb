```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
df = pd.DataFrame(df).fillna(0) 
y = df.Satisfaction
```

```python
df.head(25)
```

```python
df.drop(columns=['Satisfaction'], inplace=True)
df
```

```python
y
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.5, random_state=0, stratify=y)
```

```python
# print(len(y[y==' <=50K'])/len(y), len(y[y==' >50K']))
```

```python
y = y_train
y_str = y.astype('category')
```

```python
y_str
```

```python


for c in X_train.columns:
    if type(X_train[c].iloc[0]) == str:
        X_train[c] = X_train[c].astype('category')
        X_train[c] = X_train[c].cat.codes
```

```python
for c in X_test.columns:
    # print(X_test[c].iloc[0])
    if type(X_test[c].iloc[0]) == str:
        X_test[c] = X_test[c].astype('category')
        X_test[c] = X_test[c].cat.codes
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0, max_depth=3)
clf.fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, preds)
```

```python
accuracy
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(14,11))
plt.title('Satisfação')
tree.plot_tree(clf, label=None, node_ids=True, rounded=True, fontsize=9, feature_names=X_train.columns, filled=True, class_names=["Neutro ou Não Satisfeito", "Satisfeito"], proportion=True)
# plt.savefig('decision_tree.png', format='png')
plt.show()
```
